## Tipos de datos

```python
z = np.array(3, dtype = np.uint8)
print(z)
doublearray = np.array([1,2,3,4,5], dtype = "d") #np.float64
print(doublearray)
z = z.astype(np.float32)
print(z)
```

```python
array = np.array([[1, 2, 3], [4, 5, 6]])
sum = np.sum(array)
print(sum)
```

```python
mean = np.mean(array)
std = np.std(array)
print(std, mean)
```

## Vectores y matrices
*  Un vector puede representar una lista de valores que podrían ser coordenadas espaciales, ienen magnitud y dirección.
*  Una matriz puede representar una transformación que afecta a estos vectores.
*  Transformaciones Lineales: Son funciones que toman vectores como entrada y producen otros vectores como salida, respetando las operaciones de suma y multiplicación por un escalar.
*  Espacios Vectoriales: Conjuntos de vectores que pueden sumarse entre sí y multiplicarse por escalares, siguiendo ciertas reglas.

```python
A = np.array([[1, 2], [6, 4]])
B = np.array([[3, 5], [8, 4]])
suma = A+B
print(suma)
```

image.png

```python
producto = np.dot(A,B)
print("El producto de matrices es:/n", producto)
```
