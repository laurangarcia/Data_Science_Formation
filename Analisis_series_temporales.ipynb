```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("online_retail.csv")
print(df.info)
```

```python
df.info()
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df.info()
```

```python
# Eliminar filas con valores nulos en la columna 'InvoiceDate'
df.dropna(subset=['InvoiceDate'], inplace=True) #Inplace para que no se cree una copia, pues queremos hacer la modificacion al dataset
```

```python
df.info()
```

```python
# Establecer 'InvoiceDate' como el índice del DataFrame
df.set_index('InvoiceDate')
```

```python
df.set_index('InvoiceDate', inplace=True) 
```

```python
df['Year'] = df.index.year
df['Month'] = df.index.month
df['Day'] = df.index.day
df['Weekday'] = df.index.weekday  
df['Hour'] = df.index.hour


print(df['Year'].head())
print(df['Month'].head())
print(df['Day'].head())
print(df['Weekday'].head()) 
print(df['Hour'].head())
```
