# Uniones

```python
import pandas as pd

#Crear dataframes de ejemplo
df1 = pd.DataFrame({
    'id': [1, 2, 3],
    'nombre': ['Ana', 'Luis', 'Pedro']
})

df2 = pd.DataFrame({
    'id': [2, 3, 4],
    'nombre' : ['Luis', 'Pancho', 'María']
})

print(df1)
print(df2)
```

```python
ineer_merged = pd.merge(df1, df2, on = 'id', how = 'inner')
print(ineer_merged)
```

```python
outer_merged = pd.merge(df1, df2, on = 'id', how = 'outer')
print(outer_merged)
```

```python
left_merged = pd.merge(df1, df2, on = 'id', how = 'left')
print(left_merged)
```

```python
right_merged = pd.merge(df1, df2, on = 'id', how = 'right')
print(right_merged)
```

```python
df3 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2']
})
df4 = pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5']
})

print(df3)
print(df4)
```

```python
#Concatenar vertical axis = 0
vertical_concat = pd.concat([df3, df4], axis=0)
print(vertical_concat)
```

```python
#Concatenar horizontal axis = 1 
horizontal_concat = pd.concat([df3, df4], axis=1)
print(horizontal_concat)
```

# Joined
Combina dataframes con base a un indice o columna clave

```python
df5 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2']
}, index = ['K0', 'K1', 'K2'])
df6 = pd.DataFrame({
    'C': ['C0', 'C1', 'C2'],
    'D': ['D0', 'D1', 'D2']
}, index = ['K0', 'K2', 'K3'])

print(df5)
print(df6)
```

```python
joined = df5.join(df6, how='inner')
print(joined)
```
