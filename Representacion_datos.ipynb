```python
import matplotlib as plt
import matplotlib.pyplot as plt
import numpy as np
```

# Grafico de lineas

Los gráficos de líneas son usualmente utilizados para mostrar tendencias a lo largo del tiempo o cambios continuos en los datos.

```python
month = np.array(['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio',
                  'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre'])

sales = np.array([30, 80, 130, 150, 170, 300, 200, 120, 140, 260, 280, 300])

#Configurar el tamaño del grafico
plt.figure(figsize = (8,6))

#Crear el grafico de dispersion
plt.plot(month, sales, marker = 'o', color = 'pink')
#Configurar el titulo y etiquetas
plt.title('Ventas Mensuales', fontsize = 20)
plt.xlabel('Meses')
plt.ylabel('Ventas en miles de unidddes')
plt.show()
```

# Grafico de Dispersion
Un gráfico de dispersión es ideal para visualizar la relación entre dos variables. Por ejemplo, podrías estar interesado en saber cómo la cantidad de horas estudiadas afecta el rendimiento en un examen. De esta forma, podrías estructurar tus datos en listas

```python
#configurar el tamaño del grafico
plt.figure(figsize=(8, 6))

hours = [2, 3, 4, 5, 6, 7, 8, 9, 10]
exam = [50, 55, 60, 65, 70, 75, 80, 85, 90]

#Crear el grafico de dispersion
plt.scatter(hours, exam, color = 'pink', marker = 'o')

plt.title('Relacion horas estudiadas y su puntaje', fontsize = 20)
plt.xlabel('Horas')
plt.ylabel('Puntaje')
plt.show()
```

```python
#configurar el tamaño del grafico
plt.figure(figsize=(8, 6))

hours = [2, 3, 4, 5, 6, 7, 8, 9, 10]
exam_Juan= [50, 55, 60, 65, 70, 75, 80, 85, 90]
exam_Andres= [52, 55, 70, 85, 90, 75, 90, 85, 100]

#Crear el grafico de dispersion
plt.scatter(hours, exam_Juan, color = 'pink', marker = 'o', linewidth = 2, linestyle = '-' , label = 'Juan')
plt.scatter(hours, exam_Andres, color = 'green', marker = 's', linewidth = 2, linestyle = '-' , label = 'Andres')

plt.title('Relacion horas estudiadas y su puntaje', fontsize = 20)
plt.xlabel('Horas')
plt.ylabel('Puntaje')
plt.show()
```

## Barras



```python
plt.figure(figsize=(8, 6))

categories = ['Producto A', 'Producto B', 'Producto C', 'Producto D', 'Producto E']
sales = [10, 20, 15, 25, 30]

#Crear el grafico de barras
plt.bar(categories, sales, color='skyblue', label='Ventas mensuales')
plt.title('Ventas por Producto', fontsize=20)
plt.xlabel('Productos')
plt.ylabel('Ventas')

#Anotacion con flecha para destacar un punto
plt.annotate('Mayor venta', xy=('Producto E', 30), xytext=('Producto C', 35),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.legend()

plt.show()
```

## Diagrama de Pastel

'%1.1f%%': Muestra el porcentaje con un decimal y el símbolo %. Ejemplo: 35.7%.
'%1.0f%%': Muestra el porcentaje como un número entero sin decimales. Ejemplo: 36%.
'%1.2f%%': Muestra el porcentaje con dos decimales. Ejemplo: 35.68%.
'%2.0f%%': Similar a '%1.0f%%', pero reserva más espacio (dos dígitos enteros). Ejemplo: 35% se vería como 35%, pero 5% se vería como 5%.
'%1.1f': Muestra el porcentaje con un decimal, pero sin el símbolo %. Ejemplo: 35.7.
'%1.1f units': Personaliza el texto. En este caso, mostraría el porcentaje seguido de la palabra “units”. Ejemplo: 35.7 units.

```python

plt.figure(figsize=(8, 6))

categories = ['Producto A', 'Producto B', 'Producto C']
sales = [10, 20, 15]

#Crear el diagrama de barras
plt.pie(sales, labels = categories, autopct='%1.1f%%', startangle=140, colors=['lightcoral', 'lightskyblue', 'lightgreen'])


plt.title('Ventas por Producto')

#Asegurar que el pie chart sea un circulo
plt.axis('equal')


plt.show()
```

## Boxplots

Este gráfico proporciona una representación visual clara de la distribución de un conjunto de datos, destacando la mediana, los cuartiles y los valores atípicos.



Componentes de un Boxplot:

Box (Caja): Representa el rango intercuartil (IQR), que incluye el 50% central de los datos. Los bordes de la caja indican el primer cuartil (Q1) y el tercer cuartil (Q3).

Median (Mediana): Indica el valor central de los datos, dividiendo el conjunto en dos partes iguales.

Whiskers (Bigotes): Extienden desde los cuartiles hasta los valores mínimos y máximos dentro de 1.5 veces el IQR. Ayudan a identificar la extensión de los datos sin incluir los valores atípicos.

Outliers (Valores Atípicos): Representan los valores que se encuentran fuera del rango de los bigotes.
El boxplot es especialmente útil para comparar la distribución de los datos entre diferentes grupos o categorías y para identificar rápidamente anomalías o valores extremos en un conjunto de datos.



```python
np.random.seed(0) #la semilla de inicio hace que cada vez que jecutemos no se generen nuevos valores random

ages = [np.random.normal(30,5,100),
        np.random.normal(40,5,100),
        np.random.normal(35,5,100)]
print(ages)
```

```python
plt.boxplot(ages, patch_artist = True, notch = True, vert = True,
            labels = ['Grupo 1', 'Grupo 2', 'Grupo 3'])

plt.title('Distribucion de edades')
plt.xlabel('Grupo')
plt.ylabel('Edad (años)')
plt.show()
```

## Visualización de Series Temporales


```python
import pandas as pd


dates = pd.date_range(start = '2023-01-01', periods = 100)

values = np.random.rand(100).cumsum()

data = pd.DataFrame({'Date': dates, 'Value': values})
```

```python

fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(data['Date'], data['Value'], color = 'green')

plt.xticks(rotation = 45)

plt.title('Serie de tiempo con formato en las fechas')
plt.xlabel('Date')
plt.ylabel('Value')
```

```python
from matplotlib.dates import DateFormatter

#graficar las ventas mensuales

dates = pd.date_range(start = '2023-01-01', periods = 12, freq = 'M')
sales = np.random.randint(1000,5000, size = 12)

salesdata = pd.DataFrame=({'Date': dates, 'Sales': sales})

plt.plot(salesdata['Date'],salesdata['Sales'], marker = 'o', linestyle = '-', label = 'Ventas mensuales')

plt.gca().xaxis.set_major_formatter(DateFormatter('%b %Y'))

#Rotacion de las ettiquetas
plt.xticks(rotation = 45)

plt.title('Analisis de ventas mensuales')
plt.xlabel('Date')
plt.ylabel('Sales')

plt.legend()
plt.tight_layout()
plt.show()
```

# Gridspec
## Creacion de Graficos en una sola figura



```python
import matplotlib.pyplot as plt 
import numpy as np
import matplotlib.gridspec as gridspec

x = np.linspace(0,10, 100)
y = np.sin(x)

data = np.random.randn(100)
```

```python
gs = gridspec.GridSpec(2,2, height_ratios=[2,1], width_ratios=[1,1]) #Queremos que este dividido en 2 filas y 2 columnas 
fig = plt.figure(figsize = (10,8))

#primer subplot grande, ocupa toda la primera fila
ax1 = fig.add_subplot(gs[0,:])
ax1.plot(x,y, color = 'blue')
ax1.set_title('Seno de X')
ax1.set_xlabel('X')
ax1.set_ylabel('seno')

#segundo subplot, ocupa la esquina inferior izquierda
ax2 = fig.add_subplot(gs[1,0])
ax2.hist(data, bins =20, color = 'purple', edgecolor = 'black')
ax1.set_title('histograma')
ax1.set_xlabel('Valor')
ax1.set_ylabel('Frecuencia')

#Tercer subplot, ubicado en la esquina inferior derecha
ax3 = fig.add_subplot(gs[1,1])
ax3.scatter(x,y,color = 'red')
ax1.set_title('dispersion del seno')
ax1.set_xlabel('X')
ax1.set_ylabel('Frecuencia')


#Ajustar el layout para evitar la superposicion 
plt.tight_layout()

plt.show()
```
