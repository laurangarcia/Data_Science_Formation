```python
import matplotlib as plt
import matplotlib.pyplot as plt
import numpy as np
```

# Grafico de lineas

Los gráficos de líneas son usualmente utilizados para mostrar tendencias a lo largo del tiempo o cambios continuos en los datos.

```python
month = np.array(['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio',
                  'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre'])

sales = np.array([30, 80, 130, 150, 170, 300, 200, 120, 140, 260, 280, 300])

#Configurar el tamaño del grafico
plt.figure(figsize = (8,6))

#Crear el grafico de dispersion
plt.plot(month, sales, marker = 'o', color = 'pink')
#Configurar el titulo y etiquetas
plt.title('Ventas Mensuales', fontsize = 20)
plt.xlabel('Meses')
plt.ylabel('Ventas en miles de unidddes')
plt.show()
```

# Grafico de Dispersion
Un gráfico de dispersión es ideal para visualizar la relación entre dos variables. Por ejemplo, podrías estar interesado en saber cómo la cantidad de horas estudiadas afecta el rendimiento en un examen. De esta forma, podrías estructurar tus datos en listas

```python
#configurar el tamaño del grafico
plt.figure(figsize=(8, 6))

hours = [2, 3, 4, 5, 6, 7, 8, 9, 10]
exam = [50, 55, 60, 65, 70, 75, 80, 85, 90]

#Crear el grafico de dispersion
plt.scatter(hours, exam, color = 'pink', marker = 'o')

plt.title('Relacion horas estudiadas y su puntaje', fontsize = 20)
plt.xlabel('Horas')
plt.ylabel('Puntaje')
plt.show()
```

```python
#configurar el tamaño del grafico
plt.figure(figsize=(8, 6))

hours = [2, 3, 4, 5, 6, 7, 8, 9, 10]
exam_Juan= [50, 55, 60, 65, 70, 75, 80, 85, 90]
exam_Andres= [52, 55, 70, 85, 90, 75, 90, 85, 100]

#Crear el grafico de dispersion
plt.scatter(hours, exam_Juan, color = 'pink', marker = 'o', linewidth = 2, linestyle = '-' , label = 'Juan')
plt.scatter(hours, exam_Andres, color = 'green', marker = 's', linewidth = 2, linestyle = '-' , label = 'Andres')

plt.title('Relacion horas estudiadas y su puntaje', fontsize = 20)
plt.xlabel('Horas')
plt.ylabel('Puntaje')
plt.show()
```

```python
#configurar el tamaño del grafico
plt.figure(figsize=(8, 6))

hours = [2, 3, 4, 5, 6, 7, 8, 9, 10]
exam_Juan= [50, 55, 60, 65, 70, 75, 80, 85, 90]
exam_Andres= [52, 55, 70, 85, 90, 75, 90, 85, 100]

#Crear el grafico de dispersion
plt.plot(hours, exam_Juan, color = 'pink', marker = 'o', linewidth = 2, linestyle = '-' , label = 'Juan')
plt.plot(hours, exam_Andres, color = 'green', marker = 's', linewidth = 2, linestyle = '--' , label = 'Andres')

plt.title('Relacion horas estudiadas y su puntaje', fontsize = 20)
plt.xlabel('Horas')
plt.ylabel('Puntaje')
plt.legend()
plt.grid(True)
plt.show()
```

## Barras



```python
plt.figure(figsize=(8, 6))

categories = ['Producto A', 'Producto B', 'Producto C', 'Producto D', 'Producto E']
sales = [10, 20, 15, 25, 30]

#Crear el grafico de barras
plt.bar(categories, sales, color='skyblue', label='Ventas mensuales')
plt.title('Ventas por Producto', fontsize=20)
plt.xlabel('Productos')
plt.ylabel('Ventas')

#Anotacion con flecha para destacar un punto
plt.annotate('Mayor venta', xy=('Producto E', 30), xytext=('Producto C', 35),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.legend()

plt.show()
```

```python
#Diagrama de barras horizontal
plt.figure(figsize=(8, 6))

categories = ['Producto A', 'Producto B', 'Producto C', 'Producto D', 'Producto E']
sales = [10, 20, 15, 25, 30]

#Crear el grafico de barras
plt.barh(categories, sales, color='skyblue', label='Ventas mensuales')

plt.title('Ventas por Producto', fontsize=20)
plt.ylabel('Productos')
plt.xlabel('Ventas')

plt.legend()

plt.show()
```

## Diagrama de Pastel

'%1.1f%%': Muestra el porcentaje con un decimal y el símbolo %. Ejemplo: 35.7%.
'%1.0f%%': Muestra el porcentaje como un número entero sin decimales. Ejemplo: 36%.
'%1.2f%%': Muestra el porcentaje con dos decimales. Ejemplo: 35.68%.
'%2.0f%%': Similar a '%1.0f%%', pero reserva más espacio (dos dígitos enteros). Ejemplo: 35% se vería como 35%, pero 5% se vería como 5%.
'%1.1f': Muestra el porcentaje con un decimal, pero sin el símbolo %. Ejemplo: 35.7.
'%1.1f units': Personaliza el texto. En este caso, mostraría el porcentaje seguido de la palabra “units”. Ejemplo: 35.7 units.

```python

plt.figure(figsize=(8, 6))

categories = ['Producto A', 'Producto B', 'Producto C']
sales = [10, 20, 15]

#Crear el diagrama de barras
plt.pie(sales, labels = categories, autopct='%1.1f%%', startangle=140, colors=['lightcoral', 'lightskyblue', 'lightgreen'])


plt.title('Ventas por Producto')

#Asegurar que el pie chart sea un circulo
plt.axis('equal')


plt.show()
```
